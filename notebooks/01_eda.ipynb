```python
import torch
from torchvision import datasets, transforms
import numpy as np
import matplotlib.pyplot as plt
```

```python
transform = transforms.ToTensor()

train_dataset = datasets.CIFAR10(
    root = '../data/raw',
    train=True,
    download=True,
    transform=transform
)

test_dataset = datasets.CIFAR10(
    root = '../data/raw',
    train=False,
    download=True,
    transform=transform
)
```

```python
print("Train Samples:", len(train_dataset))
print("Test Samples:", len(test_dataset))
```

```python
class_names = train_dataset.classes
class_names
```

```python
def show_images(dataset, classes, num_images=10):
    fig, axes = plt.subplots(1, num_images, figsize=(15, 3))
    for i in range(num_images):
        image, label = dataset[i]
        axes[i].imshow(image.permute(1, 2, 0))
        axes[i].set_title(classes[label])
        axes[i].axis("off")
    plt.show()

show_images(train_dataset, class_names)
```

```python
from collections import Counter

labels = [label for _, label in train_dataset]
label_counts = Counter(labels)

for idx, count in label_counts.items():
    print(class_names[idx], count)
```

### Dataset Overview

- Training samples: 50,000
- Test samples: 10,000
- Number of classes: 10
- Classes: airplane, automobile, bird, cat, deer, dog, frog, horse, ship, truck

The dataset appears to be **balanced**, with an equal number of samples per class.

### Image Characteristics

- Image size: 32  32 pixels
- Color format: RGB

Due to the low image resolution, fine-grained details are limited. This suggests that extremely deep or complex models may not be necessary, and a simple CNN-based baseline should be sufficient as a starting point.
